```python
import pandas as pd
```

```python
dataset = pd.read_csv("50_Startups.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset, dtype = int, drop_first = True)
dataset
```

```python
dataset.columns
```

```python
Input = dataset[['R&D Spend', 'Administration', 'Marketing Spend',
       'State_Florida', 'State_New York']]
Input
```

```python
Output = dataset[['Profit']]
Output
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(Input, Output, test_size = 0.3, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeRegressor
Regressor = DecisionTreeRegressor(criterion = "friedman_mse", splitter= "best")
Regressor = Regressor.fit(X_train, Y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(Regressor)
plt.show()
```

```python
Y_pred = Regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(Y_test, Y_pred)
r_score
```

```python
import pickle
filename = "Python_program_DT_FMSE.sav"
pickle.dump (Regressor, open(filename, "wb"))
```
